```python
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_holistic = mp.solutions.holistic
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
hands = mp_hands.Hands(
    static_image_mode= False,
    max_num_hands=2,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5,
)
```

```python
#Directory()
data_dir = 'Data_'
action_list = [
    'A',
    'BA',
    'KA',
    'DA',
    'E',
    'GA',
    'HA',
    'I',
    'LA',
    'MA',
    'NA',
    'NGA',
    'O',
    'PA',
    'RA',
    'SA',
    'TA',
    'U',
    'WA',
    'YA',
    'NULL'
]
```

```python
#create directory
for action in action_list:
    try:
        path = os.path.join(data_dir, action)
        print(f'Directory Exist: {path} ')
        if os.path.exists(path):
            print(f'Directory Exist: {path} ')
        else:
            os.mkdir(path)
            print(f'Create Directory: {path} ')
    except Exception as e:
        print(e)
```

```python
cap = cv2.VideoCapture(0)
no_frame = 30
action = 'A'

pt1 = (340, 50)
pt2 = (640, 350)

pt3 = (0, 50)
pt4 = (300, 350)

while cap.isOpened(): 
    for di_r in range (30):
        cv2.waitKey(2000)

        for i in range(35):

            ret, frame = cap.read()
            frame = cv2.flip(frame, 1)
            cv2.waitKey(10)
            
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
            image.flags.writeable = False                  # Image is no longer writeable
            results = hands.process(image)                 # Make prediction
            image.flags.writeable = True                   # Image is now writeable 
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
                
            if results.multi_handedness is not None:
                res = results.multi_handedness[0]
                label = res.classification[0].label
                
                
                if label == 'Left':
                    cv2.rectangle(frame, pt3, pt4, (0, 255, 0), 2)                   
                    cropped_frame = image[0:300, 50:350]
                    cropped_frame = cv2.flip(cropped_frame, 1)
                elif label == 'Right':
                    cv2.rectangle(frame, pt1, pt2, (0, 255, 0), 2)
                    cropped_frame = image[50:350, 340:640]

                
                if i < 30:
                    path = os.path.join(data_dir, action, str(di_r) +'.'+ str(i) + '.jpg')
                    cv2.imwrite(path, cropped_frame)
                    print(f'Save: {path}')
                if i == 30:
                    break
            
            cv2.putText(frame, 'START CAPTURING: ' + str(di_r) + action , (10,19), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
            cv2.imshow('Feed', frame) 

    cap.release()
    cv2.destroyAllWindows()
```
